```python
# Mapping the credit status
credit_status_mapping = {
    'Fully Repaid': 'Closed',
    'Repaying': 'Active',
    'Disbursed': 'In Progress',
    'Disbursing&Repaying': 'Active',
    'Fully Disbursed': 'Closed',
    'Disbursing': 'Active'
}

explor1['Credit Status'] = explor1['Credit Status'].map(credit_status_mapping)
explor1['Credit Status'].unique()
```

# **Random Forest For Classification**

## Import Library

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report, mean_absolute_error, r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
RFC = explor1.copy()
# USE ONLY THE SUPPORTED COLUMN
columns_RFC = [ 'Credit Status', 'Disbursed Amount (US$)', 'Undisbursed Amount (US$)',
    'Borrower\'s Obligation (US$)', 'Due to IDA (US$)', 'Original Principal Amount (US$)']
RFC = RFC[columns_RFC]
RFC.head(5)
```

```python
label_encoder = LabelEncoder()
RFC['Credit Status'] = label_encoder.fit_transform(RFC['Credit Status'])
print(list(label_encoder.classes_))
```

```python
X_class = RFC.drop(columns=['Credit Status'])
y_class = RFC['Credit Status']
```

```python
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(
    X_class, y_class, test_size=0.2, random_state=42
)
```

```python
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_class, y_train_class)
```

```python
y_pred_class = clf.predict(X_test_class)
print(classification_report(y_test_class, y_pred_class))
```

# **Regresion for Disbursed Amount Prediction**

```python
REG = RFC.copy()
REG.head(5)
```

```python
X_reg = REG.drop(columns=['Disbursed Amount (US$)'])
y_reg = REG['Disbursed Amount (US$)']
```

```python
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)
```

```python
reg = LinearRegression()
reg.fit(X_train_reg, y_train_reg)
```

```python
y_pred_reg = reg.predict(X_test_reg)
print("Mean Absolute Error:", mean_absolute_error(y_test_reg, y_pred_reg))
print("R2 Score:", r2_score(y_test_reg, y_pred_reg))
```

# **Input all prediction result to the RFC dataset**

```python
RFC['Predicted Credit Status'] = clf.predict(X_class)
```

```python
RFC['Predicted Credit Status'] = label_encoder.inverse_transform(RFC['Predicted Credit Status'])
```

```python
RFC['Predicted Disbursed Amount (US$)'] = reg.predict(X_reg)
```

## Credit Status VS Predicted Credit Status

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create subplots: use 'xy' type for Histograms
fig = make_subplots(rows=1, cols=2, specs=[[{'type':'xy'}, {'type':'xy'}]]) # Changed 'domain' to 'xy'
fig.add_trace(go.Histogram(x=RFC["Credit Status"], name="Credit Status"), 1, 1)
fig.add_trace(go.Histogram(x=RFC["Predicted Credit Status"],name="Predicted Credit Status"), 1, 2)

fig.update_layout(height=600, width=800, title_text="Credit Status vs. Predicted Credit Status")
fig.show()
```

The predicted credit statuses are very similar, with Active (8652) being the most common, followed by In Progress (2781) and Closed (1527). The small differences between the actual and predicted values suggest that the model does a good job in predicting credit status, although there are slight variations that could be explored further to understand any trends or patterns.
